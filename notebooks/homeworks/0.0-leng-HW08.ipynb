Utilizando variables dummy, compruebe si realmente el comportamiento del PIB en función de las remesas recibidas disminuyó a partir del tercer de 2016:

```python
df4 = df.copy()
df4.columns = ["year", "S", "PIB", "REM"]
df4["Dummy"] = 1
df4.iloc[0:20, df4.columns.get_loc("Dummy")] = 0
df4["REMDummy"] = df4["Dummy"] * df4["REM"]
df4
```

```python
model_df04_01= (
    smf.ols(
        formula='PIB~REM+Dummy+(REM*Dummy)',
        data=df4
    )
    .fit()
)
model_df04_01.summary()
```

```python
model_df04_01= (
    smf.ols(
        formula='PIB~REM+Dummy+REMDummy',
        data=df4
    )
    .fit()
)
model_df04_01.summary()
```

```python
# Extract coefficients
beta_1 = model_df04_01.params['REM']
beta_3 = model_df04_01.params['REMDummy']

# Extract variance-covariance matrix
cov_matrix = model_df04_01.cov_params()

# Compute standard error for beta_1 + beta_3
var_sum = (cov_matrix.loc['REM', 'REM'] + 
           cov_matrix.loc['REMDummy', 'REMDummy'] + 
           2 * cov_matrix.loc['REM', 'REMDummy'])
std_err_sum = np.sqrt(var_sum)

int_conf_perc = 0.95

# Calcular el valor crítico z para el nivel de confianza
z_score = stats.norm.ppf(1 - (1 - int_conf_perc) / 2)

# Calcular el intervalo de confianza
conf_interval = (beta_1 + beta_3) + np.array([-z_score, z_score]) * std_err_sum
print(f"Intervalo de confianza a un {int_conf_perc*100:.2f}% en la predicción: ({conf_interval[0]:,.4f}, {conf_interval[1]:,.4f})")
```

Utilizando el modelo anterior, estime el para el PIB en el segundo trimestre de 2015, si las remesas hubieran sido de 4,000 millones de dólares y en el segundo trimestre de 2020, si las remesas fueran de fuera de 3,100 millones de dólares.

```python
# Valores de los coeficientes obtenidos del modelo ajustado
beta_0 =  model_df04_01.params['Intercept'] # Valor estimado de la constante
beta_1 =  model_df04_01.params['REM'] # Coeficiente asociado a Remesas
beta_2 =  model_df04_01.params['Dummy'] # Coeficiente asociado a Dummy
beta_3 =  model_df04_01.params['REMDummy'] # Coeficiente asociado a la interacción Remesas*Dummy

# Valores de las remesas y dummy para los diferentes trimestres
remesas_2015 = 4000  # millones de dólares para el segundo trimestre de 2015
dummy_2015 = 0  # Antes del tercer trimestre de 2016

remesas_2020 = 3100  # millones de dólares para el segundo trimestre de 2020
dummy_2020 = 1  # Después del tercer trimestre de 2016

# Estimación del PIB para el segundo trimestre de 2015
PIB_2015 = beta_0 + beta_1 * remesas_2015 + beta_2 * dummy_2015 + beta_3 * (remesas_2015 * dummy_2015)

# Estimación del PIB para el segundo trimestre de 2020
PIB_2020 = beta_0 + beta_1 * remesas_2020 + beta_2 * dummy_2020 + beta_3 * (remesas_2020 * dummy_2020)

# Resultados
print(f"Estimación del PIB para el segundo trimestre de 2015: {PIB_2015:,.2f}")
print(f"Estimación del PIB para el segundo trimestre de 2020: {PIB_2020:,.2f}")
```
